```python
print("----- LIST METHODS -----")
fruits = ["Apple", "Banana", "Mango"]

fruits.append("Orange")
print("append():", fruits)
fruits.insert(1, "Grapes")
print("insert():", fruits)
fruits.remove("Banana")
print("remove():", fruits)
fruits.sort()
print("sort():", fruits)
fruits.pop()
print("pop():", fruits)

print("\n----- TUPLE METHODS -----")
numbers = (10, 20, 30, 20, 40)

print("count():", numbers.count(20))
print("index():", numbers.index(30))
print("len():", len(numbers))
print("max():", max(numbers))
print("min():", min(numbers))

print("\n----- SET METHODS -----")
my_set = {1, 2, 3}

my_set.add(4)
print("add():", my_set)
my_set.update([5, 6])
print("update():", my_set)
my_set.remove(2)
print("remove():", my_set)
my_set.discard(10)
print("discard():", my_set)
copy_set = my_set.copy()
print("copy():", copy_set)

print("\n----- DICTIONARY METHODS -----")
student = {
    "name": "Nishit",
    "age": 20
}
print("keys():", student.keys())
print("values():", student.values())
print("items():", student.items())
print("get():", student.get("name"))
student.update({"city": "Ahmedabad"})
print("update():", student)

print("\n----- IF -----")
age = 20
if age >= 18:
    print("Eligible to vote")

print("\n----- IF ELSE -----")
number = 15
if number % 2 == 0:
    print("Even")
else:
    print("Odd")

print("\n----- IF ELIF ELSE -----")
marks = 82
if marks >= 90:
    print("Grade A")
elif marks >= 75:
    print("Grade B")
else:
    print("Grade C")

print("\n----- NESTED IF -----")
username = "admin"
password = "1234"
if username == "admin":
    if password == "1234":
        print("Login Successful")
    else:
        print("Wrong Password")
else:
    print("Invalid Username")

print("\n----- BREAK -----")
for i in range(1, 11):
    if i == 6:
        break
    print(i)

print("\n----- CONTINUE -----")
for i in range(1, 6):
    if i == 3:
        continue
    print(i)

print("\n----- PASS -----")
for i in range(1, 4):
    if i == 2:
        pass
    print(i)

print("\n----- INPUT -----")
name = input("Enter your name: ")
print("Hello,", name)

print("\n----- RANGE -----")
for i in range(5):
    print(i)

print("\n----- LEN -----")
text = "Python"
print("Length:", len(text))

print("\n----- TYPE -----")
x = 100
print(type(x))

print("\n----- FOR LOOP -----")
for i in range(1, 6):
    print("Number:", i)

print("\n----- WHILE LOOP -----")
count = 1
while count <= 5:
    print("Count:", count)
    count += 1
```

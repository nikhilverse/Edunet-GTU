```python
# Create Production Level feature
df["Production Level"] = pd.cut(
    df["Production (GWh)"],
    bins=3,
    labels=["Low", "Medium", "High"]
)

df.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
print(df.columns.tolist())
```

```python
plt.figure(figsize=(10,5))

sns.histplot(
    data=df,
    x="Production (GWh)",
    bins=30,
    kde=True
)

plt.title("Distribution of Renewable Energy Production")
plt.xlabel("Renewable Energy Production (GWh)")
plt.ylabel("Frequency")

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.boxplot(data=df, y="Production (GWh)")

plt.title("Distribution of Renewable Energy Production")
plt.ylabel("Production (GWh)")

plt.show()
```

```python
# Calculate total renewable energy production for each year
yearly = df.groupby("Year")["Production (GWh)"].sum().reset_index()

# Plot the trend
plt.figure(figsize=(12,5))

sns.lineplot(
    data=yearly,
    x="Year",
    y="Production (GWh)",
    marker="o"
)

plt.title("Renewable Energy Production Over Time")
plt.xlabel("Year")
plt.ylabel("Total Production (GWh)")

plt.grid(True)

plt.show()
```

```python
### Insight

#The line chart shows the trend of total renewable energy production over the years. It helps identify whether renewable energy generation has increased, decreased, or remained stable over time. An increasing trend indicates greater adoption of renewable energy technologies and investments.
```

```python
# Top 10 countries by average renewable energy production

top = (
    df.groupby("Country")["Production (GWh)"]
      .mean()
      .sort_values(ascending=False)
      .head(10)
)

plt.figure(figsize=(12,6))

sns.barplot(
    x=top.values,
    y=top.index,
    palette="Greens_r"
)

plt.title("Top 10 Countries by Average Renewable Energy Production")
plt.xlabel("Average Renewable Energy Production (GWh)")
plt.ylabel("Country")

plt.show()
```

```python
encoded_df = pd.get_dummies(
    df,
    columns=["Country", "Energy Type", "Production Level"],
    drop_first=True
)

encoded_df.head()
```

```python
## Encoding Categorical Features

#Categorical variables such as Country and Energy Type were converted into numerical values using One-Hot Encoding (`pd.get_dummies()`).

#This transformation is necessary because machine learning algorithms require numerical input and cannot directly process text values.
```

```python

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

numeric_cols = encoded_df.select_dtypes(include=['int64', 'float64']).columns

encoded_df[numeric_cols] = scaler.fit_transform(encoded_df[numeric_cols])

encoded_df.head()
```
